```python
#Import random module which will be used to shuffle deck 
import random 

#Create Boolean to control flow of game in while loops
playing = True

#Declare variables used to create a deck of cards and assign a numeric value to each card 
suits = ("Hearts", "Diamonds", "Spades", "Clubs")
ranks = ("Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "King", "Queen", "King", "Ace")
values = {"Two":2, "Three":3, "Four":4, "Five":5, "Six":6, "Seven":7, "Eight":8,
          "Nine":9, "Ten":10, "Jack":10, "Queen":10, "King":10, "Ace":11}
```

```python
#Create a Card class. Each Card object has suit and rank

class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
# Create a Deck object that contains Card objects
# Deck objects can be shuffled and cards can be dealt one at a time from the deck 

class Deck:
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank)) # build Card objects and add them to the list which is the deck  
                
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        single_card = self.all_cards.pop()
        return single_card
```

```python
#Create a Hand class to show the value of a players hand and that can also adjust for aces  

class Hand: 
    
    def __init__(self): 
        self.cards = []
        self.value = 0
        self.aces = 0

    def add_card(self,new_card):
        self.cards.append(new_card)
        self.value += values[new_card.rank]
        if new_card.rank == "Ace":
            self.aces += 1
        
    def adjust_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
#Create a Chips class to keep a track of a players chips and ongoing winnings 

class Chips:
    
    def __init__(self):
        self.total = 50
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
# Use a function to accept player bets 
# It takes in Chip object as an argument 

def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input("How many chips would you like to bet? "))
        except ValueError:
            print("Sorry, you must bet an integer")
        else:
            if chips.bet > chips.total:
                print("You can't bet more than your total chips which is", chips.total)
            else:
                break
```

```python
# Use a function to perform a hit
# It takes in Deck and Hand objects as arguments 
# The function can deal a card from the deck an dadd it the the Hand 

def hit(deck,hand):
    
    hand.add_card(deck.deal_one())
    hand.adjust_ace()
```

```python
# Use a function to decide on a players move 
# It takes in Deck and Hand objects as arguments 
# The function asks a player to hit or stick and will continue to do so until the player sticks

def move(deck,hand):
    
    global playing 
    
    while True:
        try:
            play = input("Would you like to hit or stay? Enter 'h' or 's':  ")
        except:
            print("Sorry, you must enter 'h' or 's' ")
        else:
            if play == "h":
                hit(deck,hand)
            else:
                print("Player stands. Dealer is playing")
                playing = False
        break
```
